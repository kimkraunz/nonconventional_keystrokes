```python
keys_to_get = ["Space", "Back", "LShiftKey", "Tab", "OemPeriod", "LShiftKey", "LMenu", "RShiftKey", "LControlKey", "Escape", "Delete"]
```

```python
cmu_005 = parse_input_data("/Users/kimkraunz/Desktop/005001.txt")
def parse_counts(data, keys_to_get, user_id):
    total_n = len(data)
    data_counts = data[data['action']=="KeyDown"].groupby("key").size().reset_index()
    data_counts.columns = ['key', 'n']
    data_counts = data_counts.sort_values(by='n', ascending=False)
    data_counts['n_per_100'] = 100*data_counts['n']/total_n
    data_counts = data_counts[data_counts['key'].isin(keys_to_get)]
    # data_counts['user'] = user_id
    return data_counts

counts_005 = parse_counts(cmu_005, keys_to_get, "cmu_005")

counts_005

```
